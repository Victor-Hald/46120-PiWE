## Task 2 - If/else statements

```python
def goldilocks (length):
    if length < 140:
        return "Too small!"
    elif length > 150:
        return "Too large!"
    else:
        return "Just right. :)"

print(goldilocks(139))
print(goldilocks(140))
print(goldilocks(151))
print(goldilocks(150))
```

## Task 3 - For loops

```python
def square_list (numbers):
    squared = []
    for sausage in numbers:
        squared.append(sausage ** 2)
    return squared

print(square_list([1, 2, 3]))
```

## Task 4 - While loops

```python
def fibonacci_stop (n):
    fib_sequence = [1, 1]
    while True:
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        if next_fib > n:
            break
        fib_sequence.append(next_fib)
    return fib_sequence

print(fibonacci_stop(30))
```

## Task 5 - Logical operators

```python
def clean_pitch (pitch, status):
    for i in range(len(pitch)):
        if status[i] == 1:
            pitch[i] = -999
    return pitch

x = [-1, 2, 6, 95]
status = [1, 0, 0, 0]
print(clean_pitch(x, status))
```

```python
# Improved
def clean_pitch_improved (pitch):
    for i in range(len(pitch)):
        if pitch[i] < 0 or pitch[i] > 90:
            pitch[i] = -999

    return pitch

x = [-1, 2, 6, 95]
print(clean_pitch_improved(x))
```

# Lecture 1 - factorial code

```python
def factorial (n):
    if n < 0:
        return f"Error: Can't take factorial of a negative number"
    elif n > 0:
        result = n
        for i in range(1, n):
            result = result * (i)
        return result
    else:
        return 1
    

print(factorial(100))
```
